```python
from collections import deque
n = 5
number = 31168
a = [35000] * (9*32000) 

aa = deque([n])
while aa:
    i = aa.popleft()
    if int(str(i)+str(n)) < len(a):
        a[int(str(i)+str(n))] = min(a[int(str(i)+str(n))], a[i]+1)
        aa.append(int(str(i)+str(n)))
    if 0 <= i//n < len(a):
        a[i//n] = min(a[i//n], a[i]+1)
        aa.append(i//n)
    if 0 <= i+n < len(a):
        a[i+n] = min(a[i+n], a[i]+1)
        aa.append(i+n)
    if 0 <= i-n < len(a):
        a[i-n] = min(a[i-n], a[i]+1)
        aa.append(i-n)
    if 0 <= i*n < len(a):
        a[i*n] = min(a[i*n], a[i]+1)
        aa.append(i*n)
    if 0 <= i + (n//n) < len(a):
        a[i+(n//n)] = min(a[i+(n//n)], a[i]+2)
        aa.append(i+(n//n))
    if 0 <= i - (n//n) < len(a):
        a[i-(n//n)] = min(a[i-(n//n)], a[i]+2)
        aa.append(i-(n//n))
print(a[number])
```
