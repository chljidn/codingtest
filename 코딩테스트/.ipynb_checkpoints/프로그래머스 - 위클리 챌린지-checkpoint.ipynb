1. 전력망을 둘로 나누기

```python
# 내가 푼 것
# node를 만들 때, 노드를 하나씩 뺀 것에 대해서 dfs를 모두 수행
def solution(n, wires):
    def dfs(node, start, visited):
        global count
        count += 1
        visited[start] = True
        for j in node[start]:
            if not visited[j]:
                dfs(node, j, visited)
    answer= 10001
    for i in range(len(wires)):
        node = [[] for i in range(n+1)]
        for j in range(len(wires)):
            if i != j:
                node[wires[j][0]].append(wires[j][1])
                node[wires[j][1]].append(wires[j][0])
        visited = [False for k in range(n+1)]
        total_count = []
        for k in range(1, len(node)):
            global count
            count = 0
            if not visited[k]:
                dfs(node, k, visited)
                total_count.append(count)
        answer = min(answer, abs(total_count[0] - total_count[1]))
    return answer
```

```python
# 다른 사람 답안2
uf = []

def find(a):
    global uf
    if uf[a] < 0: return a
    uf[a] = find(uf[a])
    return uf[a]

def merge(a, b):
    global uf
    pa = find(a)
    pb = find(b)
    if pa == pb: return
    uf[pa] += uf[pb]
    uf[pb] = pa

def solution(n, wires):
    global uf
    answer = int(1e9)
    k = len(wires)
    for i in range(k):
        uf = [-1 for _ in range(n+1)]
        tmp = [wires[x] for x in range(k) if x != i]
        for a, b in tmp: merge(a, b)
        v = [x for x in uf[1:] if x < 0]
        answer = min(answer, abs(v[0]-v[1]))

    return answer
```

```python
# 다른 사람 답안3
# 대부분 bfs로 푼 듯??
from collections import defaultdict, deque

def BFS(n, startNode, arr, check):
    cnt = 1
    que = deque([startNode])
    visited = [True for _ in range(n+1)]
    visited[startNode] = False

    while que:
        node = que.popleft()
        for target in arr[node]:
            if check[node][target] and check[target][node] and visited[target]:
                visited[target] = False
                que.append(target)
                cnt += 1

    return cnt

def solution(n, wires):
    answer = n
    arr = defaultdict(list)
    check = [[False for _ in range(n+1)] for _ in range(n+1)]

    for wire in wires:
        arr[wire[0]].append(wire[1])
        arr[wire[1]].append(wire[0])
        check[wire[0]][wire[1]] = True
        check[wire[1]][wire[0]] = True

    for i, wire in enumerate(wires):
        a,b = wire
        check[a][b] = False
        check[b][a] = False
        answer = min(answer, abs(BFS(n, a, arr, check) - BFS(n, b, arr, check)))
        check[a][b] = True
        check[b][a] = True

    return answer
```

2. 12주차

```python
# 내가 푼 답안
from itertools import permutations
def solution(k, dungeons):

    per = list(permutations(dungeons, len(dungeons)))
    answer = 0
    for i in per:
        count = 0
        a = k
        for j in i:
            if a >= j[0]:
                count += 1
                a -= j[1]
        answer= max(answer, count)
    return answer
```

3. 5주차 모음사전

```python
from itertools import product
def solution(word):
    a = ['A', 'E', 'I', 'O', 'U']
    b = []
    for i in range(1, len(a)+1):
        b.extend(list(map(''.join, product(a, repeat=i))))
    b.sort()
    return b.index(word) + 1
```

4. 아이템 줍기

```python
# 내가 푼 답안
# 알고리즘 자체는 다른 사람 것 참조
def solution(rectangle, characterX, characterY, itemX, itemY):
    a = [[0 for j in range(51*2)] for i in range(51*2)]
    for k in rectangle:
        for f in range(k[1]*2, (k[3]*2)+1):
            for x in range(k[0]*2, (k[2]*2)+1):
                if f == k[1]*2 or f == k[3]*2 or x == k[0]*2 or x == k[2]*2:
                    if a[f][x] == 0: a[f][x] = 1
                else: a[f][x] = 2

    characterX, characterY = characterY*2, characterX*2
    itemX, itemY = itemY*2, itemX*2
    v = [(0, 1), (1, 0),(0, -1), (-1, 0)]

    answer = []
    for k in v:
        visited = [[1 if a[i][j] == 0 or a[i][j] == 2 else 0 for j in range(len(a[0]))] for i in range(len(a))]
        if visited[characterX+k[0]][characterY+k[1]] != 1: 
            stack = [(characterX+k[0], characterY+k[1])]
            visited[characterX][characterY] = 1
            visited[characterX+k[0]][characterY+k[1]] = 2
        else: continue
        while stack:
            x, y = stack.pop()
            for i in v:
                if x+i[0] >= len(a) or x+i[0] <= -1 or y+i[1] >= len(a[0]) or y+i[1] <= -1:
                    continue
                    
                if visited[x+i[0]][y+i[1]] == 0:
                    if x+i[0] == itemX and y+i[1] == itemY:
                        answer.append(visited[x][y] + 1)
                        stack = []
                        break
                    visited[x+i[0]][y+i[1]] = visited[x][y] + 1
                    stack.append((x+i[0], y+i[1]))
    return (min(answer)-1)//2
```

```python
# 다른 사람 풀이 1
import itertools


def is_movable(cur_x, cur_y, next_x, next_y, rectangles):
    x, y = (cur_x + next_x) / 2, (cur_y + next_y) / 2
    is_on_any_border = any(
        (x in (x1, x2) and y1 <= y <= y2) or (y in (y1, y2) and x1 <= x <= x2)
        for x1, y1, x2, y2 in rectangles)
    is_inside_any_rect = any(
        x1 < x < x2 and y1 < y < y2 for x1, y1, x2, y2 in rectangles)
    return is_on_any_border and not is_inside_any_rect


def solution(rectangle, characterX, characterY, itemX, itemY):
    cur_pos = (characterX, characterY)
    prev_pos = None
    for dist in itertools.count():
        if cur_pos == (characterX, characterY) and prev_pos:
            perimeter = dist
            break
        elif cur_pos == (itemX, itemY):
            dist_to_item = dist
        for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            next_pos = (cur_pos[0] + dx, cur_pos[1] + dy)
            if next_pos != prev_pos and is_movable(*cur_pos, *next_pos,
                                                   rectangle):
                prev_pos, cur_pos = cur_pos, next_pos
                break
    return min(dist_to_item, perimeter - dist_to_item)
```

```python
# 다른 사람 풀이 2

# 4방향
dij = [0, -1, 0, 1, 0]
# 좌표계를 2배율로 늘려 구성
rects = [[0] * 102 for _ in range(102)]


# 모서리 판별
def is_corner(x, y):
    if not rects[y][x] == 1: return False
    for i in range(y - 1, y + 2):
        for j in range(x - 1, x + 2):
            if not rects[i][j]: return True
    return False


def solution(rectangle, characterX, characterY, itemX, itemY):
    # 사각형 찍기
    for square in rectangle:
        ix, iy, ex, ey = square
        for i in range(iy * 2, ey * 2 + 1):
            for j in range(ix * 2, ex * 2 + 1):
                rects[i][j] = 1

    # BFS
    q = [(0, 0) for _ in range(800)]
    q[0] = (characterX * 2, characterY * 2)
    rects[q[0][1]][q[0][0]] = 2  # 사각형과 구분한 visited 처리를 위해 2 더해줌

    ptr, tg = 1, 0
    while ptr != tg:
        qx, qy = q[tg][0], q[tg][1]
        for k in range(4):
            tx, ty = qx + dij[k], qy + dij[k + 1]
            if is_corner(tx, ty):
                q[ptr] = (tx, ty)
                rects[ty][tx] = rects[qy][qx] + 0.5  # 2배율이므로 거리를 0.5씩 가산
                ptr += 1
        tg += 1

    return rects[itemY * 2][itemX * 2] - 2  # 최종 계산에서 2를 차감
```

5. 퍼즐 조각 채우기(level 3)

```python
# 내가 푼 답안
def solution(game_board, table):
    for i in range(len(game_board)):
        for j in range(len(game_board[i])):
            if game_board[i][j] == 0:
                game_board[i][j] = 1
            else:
                game_board[i][j] = 0

    d_x = [0,0,1,-1]
    d_y = [1,-1,0,0]
    def dfs(x, y, table):
        if x >= len(table) or x <= -1 or y >= len(table[1]) or y <= -1:
            return

        if table[x][y] == 1:
            global a
            table[x][y] = 0
            a.append([x,y])
            for k in range(4):
                    dfs(x+d_x[k], y+d_y[k], table)

    matrix = [game_board, table]
    total_board = [] # game_board의 각 블록 매트릭스가 들어갈 리스트
    board_length = [] # total_board의 해당 블록의 칠해진 부분이 몇 칸인지 
    total_table = [] # table의 각 블록 매트릭스가 들어갈 리스트
    
    for idx, k in enumerate(matrix):
        for i in range(len(matrix[idx])):
            for j in range(len(matrix[idx])):
                if matrix[idx][i][j] == 1:
                    global a
                    a = []
                    dfs(i, j, matrix[idx])
                    min_x = min(list(map(lambda x:x[0], a)))
                    max_x = max(list(map(lambda x:x[0], a)))
                    min_y = min(list(map(lambda x:x[1], a)))
                    max_y = max(list(map(lambda x:x[1], a)))
                    for f in range(len(a)): # (0, 0) 부터 시작하는 블록 매트릭스로 만들기 위해서 
                        a[f][0] = a[f][0] - min_x # 각 조각이 (0, 0) 부터 시작하는 매트릭스에 대응하기 위해서는 최솟값만큼 빼준다.
                        a[f][1] = a[f][1] - min_y # 그렇게 되면 최솟값은 0이 되고, 나머지는 최솟값 만큼 빼준 값으로 치환된다. (최솟값 - 최솟값 = 0)
                    ff = [[0 if [m,n] in a else 1 for n in range(0, (max_y-min_y)+1)] for m in range(0, (max_x-min_x)+1)]
                    if idx == 0:
                        total_board.append(ff)
                        board_length.append(len(a))
                    else: 
                        total_table.append(ff)
    answer = 0
    for z in total_table:
        for p in range(4):
            if z in total_board:
                idx = total_board.index(z)
                del total_board[idx]
                answer += board_length.pop(idx)
                print(answer)
                break
            z = list(map(lambda x: list(x), list(zip(*z[::-1]))))
    return answer
```
