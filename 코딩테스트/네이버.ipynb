```python
from collections import deque

a, b = map(int, input().split())
idx = 1

visited = [False] * 200000
queue = deque([(0, b)])
visited[b] = True
while queue:
    idx, c = queue.popleft()
    
    if c < 0:
        continue
        
    f = 0
    for i in range(1, idx+1):
        f += i
        
    if c == a+f:
        print(idx)
        break
    
    else:
        if not visited[c-1]:
            visited[c-1] = True
            queue.append((idx+1, c-1))
        if not visited[c+1]:
            visited[c+1] = True
            queue.append((idx+1, c+1))
        if not visited[c*2]:
            visited[c*2] = True
            queue.append((idx+1, c*2))
    print(queue)
```
