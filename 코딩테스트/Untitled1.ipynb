```python
# 내가 푼 답안
from itertools import combinations
def solution(n, info):
    info_idx = list(range(len(info)-1, -1, -1))
    info_dict = {}
    for i, j in zip(info, info_idx):
        if i != 0:
            info_dict[j] = i

    c = []
    for i in range(1, n+1):
        b = list(combinations(info_idx, i))
        c.append(b)

    final_point_list = []
    point=  0
    final_comb = []
    final_point = 0
    for comb in c:
        for each_comb in comb:
            apeach_point = sum(info_dict.keys())
            rion_point = 0
            arrow_count = n
            comb_copy = list(each_comb)
            comb_dict = {}
            for p in each_comb:
                if p in info_dict.keys():
                    rion_point += p
                    apeach_point -= p
                    arrow_count -= info_dict[p]+1
                    comb_dict[p] = info_dict[p]+1
                else:
                    rion_point += p
                    arrow_count -= 1
                    comb_dict[p] = 1
            if arrow_count < 0: continue
            elif arrow_count > 0:
                comb_copy.append(min(set(range(0, 11))-set(each_comb)))
                comb_dict[min(set(range(0, 11))-set(each_comb))] = arrow_count
            if rion_point - apeach_point > point:
                final_point_list = list(comb_copy)
                point = rion_point - apeach_point
                final_comb = comb_dict
                final_point = rion_point
            elif rion_point - apeach_point == point:
                now = [0 if i not in comb_dict else comb_dict[i] for i in range(11)]
                final = [0 if i not in final_comb else final_comb[i] for i in range(11)]
                for z, k in zip(now, final):
                    if z > k:
                        final_point_list = list(comb_copy)
                        final_comb = comb_dict
                        final_point = rion_point
                        break
                    elif z < k:
                        break
    if final_point_list == []: return [-1]
    else: 
        if point == 0:
            return [-1]
        else: return [0 if i not in final_comb else final_comb[i] for i in range(10, -1, -1)]
```

```python
# 다른 사람 풀이
def solution(n, info):
    global answer, result

    def score(ryan):
        s = 0
        for i in range(11):
            if ryan[i] == info[i] == 0:
                continue
            if ryan[i] > info[i]:
                s += 10 - i
            else:
                s -= 10 - i
        return s

    def dfs(idx, left, ryan):
        global answer, result
        if idx == -1 and left:
            return
        if left == 0:
            s = score(ryan)
            if result < s:
                answer = ryan[:]
                result = s
            return
        for i in range(left, -1, -1):
            ryan[idx] = i
            dfs(idx-1, left-i, ryan)
            ryan[idx] = 0

    answer = [0 for _ in range(11)]
    result = 0
    dfs(10, n, [0 for _ in range(11)])
    return answer if result != 0 else [-1]
```

k진수에서 소수 개수 구하기

```python
# 내가 푼 답안
def solution(n, k):
    def base(x, d):
        if d == 10: return str(x)
        answer = ''
        while x >= 1:
            answer += str(x%d)
            x = x//d
        return answer[::-1]

    def prime(x):
        if x == 1 or x == 0: return False
        for j in range(2, int(x**0.5)+1):
            if x%j == 0:
                return False
        return True
    a = base(n, k).split("0")
    answer = 0
    for i in a:
        if i.isdecimal():
            if prime(int(i)): answer+= 1
    return answer
```

주차 요금 계산

```python
from datetime import timedelta
import math 
def solution(fees, records):
    base_time, base_fee, unit_time, unit_fee = fees 
    records_dict = {}
    answer = {}
    for i in records:
        a = i.split()
        if a[-1] == "IN":
            records_dict[a[1]] = a[0]
            answer[a[1]] = answer.get(a[1], 0)
        else:
            start = records_dict.pop(a[1])
            end = a[0]
            start = timedelta(hours = int(start[:2]), minutes=int(start[3:]))
            end = timedelta(hours = int(end[:2]), minutes=int(end[3:]))
            fee = ((end-start).seconds)//60
            answer[a[1]] += fee
    if records_dict:
        for i in records_dict:
            start = records_dict[i]
            end = "23:59"
            start = timedelta(hours = int(start[:2]), minutes=int(start[3:]))
            end = timedelta(hours = int(end[:2]), minutes=int(end[3:]))
            fee = ((end-start).seconds)//60
            answer[i] += fee
    final_answer = []
    for i in sorted(answer.keys()):
        if answer[i] < base_time:
            final_answer.append(base_fee)
        else:
            final_answer.append(base_fee + (math.ceil((answer[i] - base_time)/unit_time) * unit_fee))
    return final_answer
```
